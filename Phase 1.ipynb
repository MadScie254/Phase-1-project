### Analyzing Relationship Between Production Budget and Domestic Gross

```python
import re

# Filter out rows with missing values in 'production_budget' and 'domestic_gross' columns
filtered_data = data[['production_budget', 'domestic_gross']].dropna()

# Remove commas and dollar sign from the numeric values
filtered_data['production_budget'] = filtered_data['production_budget'].apply(lambda x: re.sub(r'[\$,]', '', x))
filtered_data['domestic_gross'] = filtered_data['domestic_gross'].apply(lambda x: re.sub(r'[\$,]', '', x))

# Convert the values to int
filtered_data['production_budget'] = filtered_data['production_budget'].astype(int)
filtered_data['domestic_gross'] = filtered_data['domestic_gross'].astype(int)

# Calculate the profit as the difference between domestic gross and production budget
filtered_data['profit'] = filtered_data['domestic_gross'] - filtered_data['production_budget']

# Calculate the average profit by production budget range
budget_ranges = pd.cut(filtered_data['production_budget'], bins=[0, 50000000, 100000000, float('inf')],
                       labels=['Low', 'Moderate', 'High'])
average_profit_by_budget = filtered_data.groupby(budget_ranges)['profit'].mean()

# Display the relationship between production budget range and average profit
average_profit_by_budget.plot(kind='bar', xlabel='Production Budget Range', ylabel='Average Profit')
```

### Analyzing Relationship Between Release Month and Domestic Gross

```python
import re

# Filter out rows with missing or non-numeric values in 'release_date' and 'domestic_gross' columns
filtered_data = data[['release_date', 'domestic_gross']].dropna()
filtered_data['domestic_gross'] = filtered_data['domestic_gross'].apply(lambda x: re.sub(r'[\$,]', '', x))

# Convert 'release_date' column to datetime
filtered_data['release_date'] = pd.to_datetime(filtered_data['release_date'])

# Extract the release month from the 'release_date' column
filtered_data['release_month'] = filtered_data['release_date'].dt.month

# Convert 'domestic_gross' column to numeric
filtered_data['domestic_gross'] = pd.to_numeric(filtered_data['domestic_gross'])

# Calculate the total domestic gross by release month
total_domestic_gross = filtered_data.groupby('release_month')['domestic_gross'].sum()

# Display the relationship between release month and total domestic gross
total_domestic_gross.plot(kind='bar', xlabel='Release Month', ylabel='Total Domestic Gross')
```

### Recommendation 3: Relationship between the production budget and the worldwide gross sales

```python
import matplotlib.pyplot as plt

# Filter out rows with missing or non-numeric values in 'production_budget' and 'worldwide_gross' columns
filtered_data = data[['production_budget', 'worldwide_gross']].dropna()

# Convert the values to numeric
filtered_data['production_budget'] = filtered_data['production_budget'].str.replace(',', '').str.replace('$', '').astype(float)
filtered_data['worldwide_gross'] = filtered_data['worldwide_gross'].str.replace(',', '').str.replace('$', '').astype(float)

# Calculate the profit as the difference between worldwide gross and production budget
filtered_data['profit'] = filtered_data['worldwide_gross'] - filtered_data['production_budget']

# Calculate the return on investment (ROI) as the profit divided by the production budget
filtered_data['ROI'] = filtered_data['profit'] / filtered_data['production_budget']

# Sort the data by ROI in descending order
sorted_data = filtered_data.sort_values('ROI', ascending=False)

# Select the top 10 movies with the highest ROI
top_movies = sorted_data.head(10)

# Display the top movies and their production budget, worldwide gross, and ROI
top_movies[['production_budget', 'worldwide_gross', 'ROI']]
```

### Visualization 1: Box Plot - Distribution of Worldwide Gross by Release Month


```python
import seaborn as sns
import calendar

# Extract month from the release date
data['release_month'] = pd.to_datetime(data['release_date']).dt.month

# Create a box plot to show the distribution of worldwide gross by release month
plt.figure(figsize=(10, 6))
sns.boxplot(x='release_month', y='worldwide_gross', data=data)
plt.xlabel('Release Month')
plt.ylabel('Worldwide Gross')
plt.title('Distribution of Worldwide Gross by Release Month')
plt.xticks(range(1, 13), calendar.month_abbr[1:13])
plt.show()
```

### Visualization 2: Scatter Plot - Relationship between Production Budget and Worldwide Gross


```python
plt.figure(figsize=(8, 6))
plt.scatter(data['production_budget'], data['worldwide_gross'], alpha=0.5)
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.title('Relationship between Production Budget and Worldwide Gross')
plt.show()
```

### Visualization 3: Bar Plot - Top 10 Movies by Worldwide Gross


```python
# Convert 'worldwide_gross' column to numeric
data['worldwide_gross'] = data['worldwide_gross'].str.replace(',', '').str.replace('$', '').astype(float)

# Select the top 10 movies with the highest worldwide gross
top_10_movies = data.nlargest(10, 'worldwide_gross')

plt.figure(figsize=(10, 6))
sns.barplot(x='worldwide_gross', y='movie', data=top_10_movies, palette='viridis')
plt.xlabel('Worldwide Gross')
plt.ylabel('Movie')
plt.title('Top 10 Movies by Worldwide Gross')
plt.show()
```
